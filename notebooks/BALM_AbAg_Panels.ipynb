## Panel 7 | Predicted vs measured scatters (reads raw data files)

```python
from scipy.stats import pearsonr, spearmanr
import glob
def best_fold(d,fcol,tcol,pcol):
    bf=max(d[fcol].unique(),key=lambda fi:pearsonr(d[d[fcol]==fi][tcol],d[d[fcol]==fi][pcol])[0]); return d[d[fcol]==bf]
def scat(ax,t,p,title,lab):
    t=np.asarray(t,float); p=np.asarray(p,float); r=pearsonr(t,p)[0]; rho=spearmanr(t,p)[0]; rmse=float(np.sqrt(np.mean((t-p)**2)))
    xs=np.linspace(t.min(),t.max(),100); rng=np.random.default_rng(0); B=400; L=np.zeros((B,100))
    for b in range(B):
        idx=rng.integers(0,len(t),len(t)); m,c=np.polyfit(t[idx],p[idx],1); L[b]=m*xs+c
    ax.scatter(t,p,c=OURS,alpha=0.35,s=8,linewidths=0,rasterized=True)
    ax.plot(xs,L.mean(0),color='#7b241c',lw=1.2); ax.fill_between(xs,np.percentile(L,2.5,0),np.percentile(L,97.5,0),color=OURS,alpha=0.25)
    ax.set_xlabel(f'Measured {PKD}'); ax.set_ylabel(f'Predicted {PKD}')
    ax.set_title(f'{title}\nr={r:.3f}, ρ={rho:.3f}, RMSE={rmse:.2f} (n={len(t)})',fontsize=6.6); panel(ax,lab)
fig,axes=plt.subplots(2,3,figsize=(7.4,5.0)); axes=axes.flatten()
sa=best_fold(pd.read_csv(os.path.join(HERE,'results_saaintdb_allcdr/random/all_preds.csv')),'fold','binding_affinity','predicted_affinity')
scat(axes[0],sa['binding_affinity'],sa['predicted_affinity'],'SAaIntDB',' a')
for ax,ds,nm,lab in zip(axes[1:4],['sabdab','abbind','skempi'],['SAbDab','AB-Bind','SKEMPI'],[' b',' c',' d']):
    d=best_fold(pd.read_csv(os.path.join(HERE,f'experiments/results_allcdr_stats/oof_preds_{ds}.csv')),'fold','true','pred'); scat(ax,d['true'],d['pred'],nm,lab)
bps=glob.glob(os.path.join(HERE,'experiments/results/ours_allcdr_benchmark/seed_*/benchmark_preds.csv'))
bb=max(bps,key=lambda fp:pearsonr(pd.read_csv(fp)['true'],pd.read_csv(fp)['pred'])[0]); bd=pd.read_csv(bb)
scat(axes[4],bd['true'],bd['pred'],'Benchmark',' e')
axes[5].axis('off')
fig.suptitle('Predicted vs measured affinity (best representative fold/seed) with 95% CI band\nSAaIntDB (a), SAbDab (b), AB-Bind (c), SKEMPI (d), Benchmark (e)',fontsize=8.5,fontweight='bold')
plt.tight_layout(rect=[0,0,1,0.93]); save(fig,'panel7_scatters'); plt.show()
```
